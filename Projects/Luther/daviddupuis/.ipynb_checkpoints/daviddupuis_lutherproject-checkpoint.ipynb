Get all the information of a studio for a specific year (between 2016 - 2000)

```python
def get_all_studio_info_year(studios_list,year,studio_dict={}):
    studio_info_url = 'http://www.boxofficemojo.com/studio/?view=company&view2=yearly&yr=%s&p=.htm'
    url = studio_info_url % year
    print(url)
    soup = get_url_soup_content(url)
    
    list_length = len(soup.find_all('table')[2].find_all('tr')) - 2 #correct table length
    for i in range(2,list_length):
        studio_name = str(soup.find_all('table')[2].find_all('tr')[i].find_all('td')[1].text) #studio name
        
        if studio_name in studios_list:
            studio_url = str(soup.find_all('table')[2].find_all('tr')[i].find_all('td')[1].find('a')['href']) #studio url
            url_name = re.search('[\w]+.htm',studio_url).group(0).replace('.htm','')
            studio_marketShare = float(soup.find_all('table')[2].find_all('tr')[i].find_all('td')[2].text.strip('%')) #studio market share
            '''
            total gross can be either in millions (*) or in thousands (ending with K)
            '''
            gross_string = soup.find_all('table')[2].find_all('tr')[i].find_all('td')[3].text.strip('$').replace(',','')
            if gross_string[len(gross_string)-1] == 'k':
                gross_string = gross_string.replace('k','') + 'E+3'
                studio_totalGross = float(gross_string)
            else: 
                gross_string = gross_string + 'E+6'
                studio_totalGross = float(gross_string)
            studio_moviesTracked = int(soup.find_all('table')[2].find_all('tr')[i].find_all('td')[4].text) #studio movies tracked
            studio_tracked_thisYear = int(soup.find_all('table')[2].find_all('tr')[i].find_all('td')[5].text) #studio movies tracked this year
            if studio_name not in studio_dict:
                studio_dict[studio_name] = {'url_name': url_name,
                                             year:{
                                                'market_share':studio_marketShare,
                                                'total_gross':studio_totalGross,
                                                'movies_tracked':studio_moviesTracked,
                                                'movies_tracked_this_year':studio_tracked_thisYear,
                                                'movies':''}}
            else:
                studio_dict[studio_name][year]= {
                                                'market_share':studio_marketShare,
                                                'total_gross':studio_totalGross,
                                                'movies_tracked':studio_moviesTracked,
                                                'movies_tracked_this_year':studio_tracked_thisYear,
                                                'movies':''}
    return studio_dict
            
```

###The only useful functions for my Luther project

Get the content of a web page.

```python
def get_url_soup_content(url):
    response = requests.get(url)
    if response.ok:
        return BeautifulSoup(response.text)
```

Get all the studio info for a given year and a given quarter

```python
def get_studio_movies_info_by_quarter(key,quarter_url,super_dict={}):
    movie_url = 'http://www.boxofficemojo.com/movies/?id=%s.htm'
    print('>studio: ' + quarter_url)
    soup = get_url_soup_content(quarter_url)
    
    '''Initialize variables that need to be incremented'''
    studio_totalTheaters = 0
    studio_openTheaters = 0
    studio_budget = 0
    studio_openGross = 0
    movie_list = soup.find_all('table')[3].find_all('tr')

    movie_list_length = len(movie_list) - 5 # Number of movies
    
    '''Special case if there are no movies for this quarter'''
    if(movie_list_length <= 0):
        super_dict['total_gross'][key]         = 0
        super_dict['number_movies'][key]       = 0
        super_dict['total_theaters'][key]      = 0
        super_dict['total_open_theaters'][key] = 0
        super_dict['total_open_gross'][key]    = 0
        super_dict['average_gross'][key]       = 0
        super_dict['total_budget'][key]        = 0
        
        return super_dict
    
    studio_totalGross = movie_list[len(movie_list) - 3].find_all('td')[1].text.replace('$','').replace(',','') # total gross for quarter
    studio_avGross = movie_list[len(movie_list) - 2].find_all('td')[1].text.replace('$','').replace(',','') # average gross for quarter
    
    '''Special case if there is only one movie'''
    if(movie_list_length == 1):
        row = movie_list[1].find_all('td')
        
        '''soup_movie_url is the url id for a movie: "Oz the great and magnificent" = "oz"'''
        soup_movie_url = row[1].find('a')['href']
        movie_id = re.search('[\w]+.htm',soup_movie_url).group(0).replace('.htm','') # Movie url id, use to get budget

        if row[4].text != 'N/A':
            studio_totalTheaters = studio_totalTheaters + int(row[4].text.replace(',','')) # total number of theaters
        
        if row[5].text != 'N/A':
            studio_openGross = studio_openGross + int(row[5].text.replace('$','').replace(',','')) # total opening gross
        
        if row[6].text != 'N/A':
            studio_openTheaters = studio_openTheaters + int(row[6].text.replace(',','')) # total number of opening theaters
        
        '''Get the budget for each movie for this studio the quarter of this year'''
        mov_url = movie_url % movie_id
        print('>>>movie: ' + mov_url)
        soup2 = get_url_soup_content(mov_url)
        if soup2 is not None:
            temp_budget = soup2.find(text='Production Budget: ').parent.b.text

            if temp_budget != 'N/A':
                temp_budget = temp_budget.replace('$','').replace(',','').replace('million','').strip(' ') + 'E+6'
                temp_budget_int = float(temp_budget)
                studio_budget += temp_budget_int
            
    for i in range(1,movie_list_length):
        movie_url = 'http://www.boxofficemojo.com/movies/?id=%s.htm'
        row = movie_list[i].find_all('td')
        soup_movie_url = row[1].find('a')['href']
        movie_id = re.search('[-:_/\\()\w]+.htm',soup_movie_url).group(0).replace('.htm','') # Movie url id, use to get budget

        if row[4].text != 'N/A':
            studio_totalTheaters = studio_totalTheaters + int(row[4].text.replace(',','')) # total number of theaters
        
        if row[5].text != 'N/A':
            studio_openGross = studio_openGross + int(row[5].text.replace('$','').replace(',','')) # total opening gross
        
        if row[6].text != 'N/A':
            studio_openTheaters = studio_openTheaters + int(row[6].text.replace(',','')) # total number of opening theaters
        
        mov_url = movie_url % movie_id
        #print('>>>movie: ' + mov_url)
        soup2 = get_url_soup_content(mov_url)
        if soup2 is not None:
            temp_budget = soup2.find(text='Production Budget: ').parent.b.text

            if temp_budget != 'N/A' and temp_budget != 'n/a':
                temp_budget = temp_budget.replace('$','').replace(',','').replace('million','').strip(' ') + 'E+6'
                temp_budget_int = float(temp_budget)
                studio_budget += temp_budget_int
        
    #movie_releaseDate = row[7].text
    
    super_dict['total_gross'][key]         = int(studio_totalGross)
    super_dict['number_movies'][key]       = movie_list_length
    super_dict['total_theaters'][key]      = studio_totalTheaters
    super_dict['total_open_theaters'][key] = studio_openTheaters
    super_dict['total_open_gross'][key]    = studio_openGross
    super_dict['average_gross'][key]       = int(studio_avGross)
    super_dict['total_budget'][key]        = studio_budget
    
    return super_dict
        
```

Get for each quarter of each year information:
- number of movies released
- total gross for studio
- total gross for each movie
- total budget for movies (if available)
- average gross
- total theaters
- total opening gross
- total opening theaters

```python
def get_studio_quarter_info(studio_url_name='buenavista'):
    quarter_url = 'http://www.boxofficemojo.com/studio/chart/?view=company&view2=release&yr=%s&timeframe=%s&sort=&order=&studio=%s.htm'
    
    quarters = ['Q1','Q2','Q3','Q4']
    data_dict = {'total_gross':{},
               'number_movies':{},
               'total_theaters':{},
               'total_open_theaters':{},
               'total_open_gross':{},
               'average_gross':{},
               'total_budget':{}} 
    year = 2015
    while year >= 1981:#go to 1981
        for quarter in quarters:
            #get information for this quarter
            dic_key = str(year) + quarter
            url = quarter_url % (year, quarter, studio_url_name) #year, quarter, studio_url
            data_dict = get_studio_movies_info_by_quarter(dic_key,url,data_dict)
            
        year -= 1
    return data_dict
```

Making ols model with only total_open_gross as parameter to study direct correlation with total_gross. And generating plot to visualize linear regression model between both variables.

```python
bv_lr1 = LinearRegression()
bv_X = copy_bv_df[['number_movies','total_budget','total_open_gross','total_theaters']]
bv_y = copy_bv_df[['total_gross']]
```

```python
bv_X.head()
```

```python
x = []
y = []
for i in range(0,101):
    X_train, X_test, y_train, y_test = train_test_split(bv_X, bv_y, test_size=0.2)
    bv_lr1.fit(X_train,y_train)
    y.append(bv_lr1.score(X_test,y_test))
    x.append(i)
    if i >= 100:
        print(bv_lr1.score(X_test,y_test))
plt.xlabel("iterations")
plt.ylabel("test_size")
print(plt.plot(x,y))
```
